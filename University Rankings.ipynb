# Database ConnectionðŸ’¾

```python
with sqlite3.connect('university rankings.db') as conn:
    df.to_sql('rankings', conn, if_exists='replace', index=False)
```

# Data CleaningðŸ“Œ

```python
query = """
UPDATE rankings
SET
    RANK_2024 = COALESCE(RANK_2024, 0),
    STATUS = COALESCE(STATUS, "Unknown"),
    International_Faculty_Score = COALESCE(International_Faculty_Score, 0),
    International_Faculty_Rank = COALESCE(International_Faculty_Rank, 0),
    International_Students_Score = COALESCE(International_Students_Score, 0),
    International_Students_Rank = COALESCE(International_Students_Rank, 0),
    International_Research_Network_Score = COALESCE(International_Research_Network_Score, 0),
    International_Research_Network_Rank = COALESCE(International_Research_Network_Rank, 0),
    Sustainability_Score = COALESCE(Sustainability_Score, 0),
    Sustainability_Rank = COALESCE(Sustainability_Rank, 0),
    Overall_Score = COALESCE(Overall_Score, 0)
WHERE
    RANK_2024 IS NULL
    OR STATUS IS NULL
    OR International_Faculty_Score IS NULL
    OR International_Faculty_Rank IS NULL
    OR International_Students_Score IS NULL
    OR International_Students_Rank IS NULL
    OR International_Research_Network_Score IS NULL
    OR International_Research_Network_Rank IS NULL
    OR Sustainability_Score IS NULL
    OR Sustainability_Rank IS NULL
    OR Overall_Score IS NULL
;   
"""

conn.execute(query)
conn.commit()
```

```python
query = """
UPDATE rankings
SET 
    RANK_2025 = CASE
        WHEN RANK_2025 LIKE '%-%' THEN SUBSTR(RANK_2025, 1, INSTR(RANK_2025, '-') - 1)
        ELSE REPLACE(REPLACE(RANK_2025, '+', ''), '=', '')
    END,
    
    RANK_2024 = CASE
        WHEN RANK_2024 LIKE '%-%' THEN SUBSTR(RANK_2024, 1, INSTR(RANK_2024, '-') - 1)
        ELSE REPLACE(REPLACE(RANK_2024, '+', ''), '=', '')
    END,
    
    Region = REPLACE(Region, 'Not Classified', 'Europe'),
    
    Academic_Reputation_Rank = REPLACE(Academic_Reputation_Rank, '+', ''),
    
    Employer_Reputation_Rank = REPLACE(Employer_Reputation_Rank, '+', ''),
    
    Faculty_Student_Rank = REPLACE(Faculty_Student_Rank, '+', ''),
    
    Citations_per_Faculty_Rank = REPLACE(Citations_per_Faculty_Rank, '+', ''),
    
    International_Faculty_Rank = REPLACE(International_Faculty_Rank, '+', ''),
    
    International_Students_Rank = REPLACE(International_Students_Rank, '+', ''),
    
    International_Research_Network_Rank = REPLACE(International_Research_Network_Rank, '+', ''),
    
    Employment_Outcomes_Rank = REPLACE(Employment_Outcomes_Rank, '+', ''),

    Sustainability_Rank = CASE
        WHEN Sustainability_Rank LIKE '%-%' THEN SUBSTR(Sustainability_Rank, 1, INSTR(Sustainability_Rank, '-') - 1)
        ELSE REPLACE(REPLACE(Sustainability_Rank, '+', ''), '=', '')
    END
;
"""

conn.execute(query)
conn.commit()
```

# Regional Performance Comparison

```python
query3 = """
SELECT SUM(Overall_Score) AS Overall_Score, Region
FROM rankings
WHERE Region IS NOT NULL
GROUP BY Region
ORDER BY Overall_Score DESC
LIMIT 5
;
"""

result3 = pd.read_sql(query3, conn)
result3
```

```python
figure = px.bar(result3, x='Region', y='Overall_Score', title='Regional Performance Comparison', labels={'Overall_Score': 'Overall Score'})
figure.show()
```

# Top 3 Universities per Region

```python
query4 = """
SELECT Institution_Name, Region, Overall_Score
FROM (
    SELECT
        *,
        ROW_NUMBER() OVER (PARTITION BY Region ORDER BY Overall_Score DESC) AS rank_per_region
    FROM rankings
)
WHERE rank_per_region <= 3
ORDER BY Region, rank_per_region
;
"""

result4 = pd.read_sql(query4, conn)
result4
```

# Number of Universities per Region

```python
query5 = """
SELECT Region, COUNT(Institution_Name) AS Number_of_Universities
FROM rankings
GROUP BY Region
ORDER BY Number_of_Universities DESC
;
"""

result5 = pd.read_sql(query5, conn)
result5
```

# Average Overall Score per Region

```python
query6 = """
SELECT Region, AVG(Overall_Score) AS Avg_Overall_Score
FROM rankings
GROUP BY Region
ORDER BY Avg_Overall_Score DESC
;
"""

result6 = pd.read_sql(query6, conn)
result6
```

# Region Ranking Based on Average Overall Score

```python
query7 = """
SELECT Region, AVG(Overall_Score) AS Avg_Overall_Score,
    RANK() OVER (ORDER BY AVG(Overall_Score) DESC) AS Rank
FROM rankings
GROUP BY Region
ORDER BY Rank
;
"""

result7 = pd.read_sql(query7, conn)
result7
```

# The Best 7 Universities in Indonesia

```python
query10 = """
WITH AsiaRanks AS (
    SELECT 
        Institution_Name,
        ROW_NUMBER() OVER (ORDER BY Overall_Score DESC) AS Asia_Rank
    FROM rankings
    WHERE Region = 'Asia'
)

SELECT 
    indo.Institution_Name,
    indo.Overall_Score,
    ROW_NUMBER() OVER (ORDER BY indo.Overall_Score DESC) AS Indonesia_Rank,
    indo.RANK_2025 AS World_Rank,
    asia.Asia_Rank
FROM rankings indo
LEFT JOIN AsiaRanks asia
    ON indo.Institution_Name = asia.Institution_Name
WHERE indo.Location = 'Indonesia'
ORDER BY indo.Overall_Score DESC
LIMIT 7
;
"""
result10 = pd.read_sql(query10, conn)
result10
```
